```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\ofek3\OneDrive\Desktop\Idea_1973_Code\data\nberces5818v1_n2012.csv")

df.head()

df.columns

# יצירת משתנים ריאליים על ידי חלוקה במדדי המחירים (Deflators)
df['real_energy'] = df['energy'] / df['pien']
df['real_vship'] = df['vship'] / df['piship']

# בדיקה קצרה של העמודות החדשות
df[['year', 'energy', 'real_energy', 'vship', 'real_vship']].head()

# חישוב היחס בין אנרגיה ריאלית לתפוקה ריאלית
df['real_energy_intensity'] = df['real_energy'] / df['real_vship']

# הערה: עדיף להשתמש בקו תחתון (_) בשמות עמודות במקום רווח
df[['year', 'naics', 'real_energy_intensity']].head()

missing_count = df['real_energy_intensity'].isnull().sum()
missing_count

# 1. הסרת שורות שבהן חסר מידע קריטי (כמו TFP או עצימות אנרגיה)
df_clean = df.dropna(subset=['real_energy_intensity', 'tfp5']).copy()

# 2. סינון השנים המדויקות למחקר: לפני המשבר, המשבר, וההתאוששות (1960-1990)
df_final = df_clean[(df_clean['year'] >= 1960) & (df_clean['year'] <= 1990)]

# בדיקה כמה נתונים נשארו לנו בסט הסופי
df_final.shape
```
